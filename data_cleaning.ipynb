```python
# Bring in dependencies 
import pandas as pd
```

```python
# Bring in both csv files for merging
main_data = pd.read_csv('cleaned_data.csv')
coordinate_data = pd.read_csv('country_coordinates.csv')
coordinate_data.head()
```

```python
main_data.head()
```

```python
# merge the two data sets into one
full_data = main_data.merge(coordinate_data, on = 'country')
full_data.head()
```

```python
# groupby country name and year in order to get total
sumed_data = full_data.groupby(['country','year']).sum()
sumed_data.head()
```

```python
# Drop all the unneccessary columns
suicide_data = sumed_data.loc[:, ["suicides_no",'suicides_per_100k','population']]
suicide_data = suicide_data.reset_index()

# change the names of the columns to be merged back into the full_data set
suicide_data = suicide_data.rename(columns={'suicides_no':'total_suicides_no','suicides_per_100k':'total_suicides_per_100k',\
                                           'population':'total_population'})

suicide_data
```

```python
# merge the new data back onto the full data set
original_data = full_data.merge(suicide_data, on = ['country','year'])
original_data_grouped = original_data.groupby(['country','year'])
original_data_grouped.mean()
```

```python
# Remaking the data because the other one did not contain enough relevant data
happy_data_2015 = pd.read_csv('2015.csv')
happy_data_2016 = pd.read_csv('2016.csv')
sui_data = pd.read_csv('suicide_rates.csv')
happy_data_2016.head()

# merge data into a dataset containing sui data for both sexes (total)
# total_sui_data = sui_data[sui_data['Sex'] == 'Both sexes']
total_sui_data = sui_data.loc[sui_data['Sex']==' Both sexes']
total_sui_data.head()

# Change all the happyness data columns to contain the year
happy_data = happy_data_2015.merge(happy_data_2016, on = 'Country', suffixes=(' 2015',' 2016'))
happy_data = happy_data.drop(columns={'Region 2016','Year 2015','Year 2016','Lower Confidence Interval',\
                                      'Upper Confidence Interval'})
happy_data = happy_data.rename(columns={'Region 2015':'Region'})

# change column names
happy_data[happy_data['Country'] == 'United States']
```

```python
total_sui_data[total_sui_data['Country'] == 'United States of America']

total_sui_data = total_sui_data.replace('United States of America', 'United States')
```

```python
# merge all the data together into one dataset
working_data = total_sui_data.merge(happy_data, on='Country', how = 'inner')
working_data

# rename columns to be better
working_data = working_data.rename(columns={'2016':'sui_per_100k_2016','2015':'sui_per_100k_2015',\
                                           '2010':'sui_per_100k_2010','2000':'sui_per_100k_2000'})
```

```python
# change columns to be easier to query later 
working_data.columns = map(str.lower, working_data.columns)
working_data.columns = working_data.columns.str.replace(' ','_')
working_data.columns = working_data.columns.str.replace('(','')
working_data.columns = working_data.columns.str.replace(')','')
working_data.head()
```

```python
working_data[working_data['country'] == 'United States']
```

```python
# merge in population 
country_data = working_data.merge(suicide_data, on = 'country', how='left')
country_data.columns

# drop the old dumpy columns we don't need
country_data = country_data.drop(columns = {'year','total_suicides_no','total_suicides_per_100k'})

country_data
```

```python
# Create new columns for total suicides and total gdp
country_data['total_suicides_2016'] = country_data['sui_per_100k_2016'] * (country_data['total_population']/100000)
country_data['total_suicides_2015'] = country_data['sui_per_100k_2015'] * (country_data['total_population']/100000)
# country_data['2010_total_suicides'] = country_data['2010_sui_per_100k'] * (country_data['total_population']/100000)
# country_data['2000_total_suicides'] = country_data['2000_sui_per_100k'] * (country_data['total_population']/100000)
country_data

# do the same but for gdp
country_data['economy_gdp_2016'] = country_data['economy_gdp_per_capita_2016'] * (country_data['total_population'])
country_data['economy_gdp_2015'] = country_data['economy_gdp_per_capita_2015'] * (country_data['total_population'])
country_data
```

```python
# add in the coordinate data
country_data = country_data.merge(coordinate_data, on = 'country', how = 'left')
country_data
```

```python
coordinate_data[coordinate_data['country'] == 'United States']
```

```python
# bringing in more data to make visualizations more robust
hdi_data = pd.read_csv('human_development.csv')
hdi_data.columns = map(str.lower, hdi_data.columns)
hdi_data.columns = hdi_data.columns.str.replace(' ','_')
hdi_data.columns = hdi_data.columns.str.replace('(','')
hdi_data.columns = hdi_data.columns.str.replace(')','')
hdi_data.head()
```

```python
hdi_data = hdi_data.rename(columns={'human_development_index_hdi':'human_development_index'})
```

```python
# merge the hdi data into country data
country_data = country_data.merge(hdi_data, on = 'country', how = 'inner')
country_data
```

```python
# drop duplicates from the country column
country_data = country_data.drop_duplicates(subset=['country'])
country_data[country_data['country'] == 'Armenia']
country_data

# round all the data to 2 decimal places
country_data = country_data.round(2)
```

```python
# Change 'United States' back to 'United States of America'
country_data = country_data.replace('United States', 'United States of America')
```

```python
country_data.to_csv(f'country_data.csv', index=False)
```

```python
# Creating another csv file while grouping by region instead of country, mean should get us the data we need
region_data = country_data.groupby('region').mean()
region_data = region_data.drop(columns = {'sui_per_100k_2016', 'sui_per_100k_2010',\
       'sui_per_100k_2000', 'happiness_rank_2015', \
       'standard_error', 'family_2015',\
       'health_life_expectancy_2015', 'freedom_2015',\
       'trust_government_corruption_2015', 'generosity_2015',\
       'dystopia_residual_2015', 'happiness_rank_2016', 'happiness_score_2016',\
       'economy_gdp_per_capita_2016', 'family_2016',\
       'health_life_expectancy_2016', 'freedom_2016',\
       'trust_government_corruption_2016', 'generosity_2016',\
       'dystopia_residual_2016', 'total_population', 'total_suicides_2016',\
       'total_suicides_2015', 'economy_gdp_2016', 'economy_gdp_2015',\
       'latitude', 'longitude', 'hdi_rank',\
       'life_expectancy_at_birth', 'expected_years_of_education',\
       'mean_years_of_education', 'gni_per_capita_rank_minus_hdi_rank'})

# export it so it can be used for analysis
region_data = region_data.reset_index()
```

```python
# round all the region data to two decimal places for better viewing
region_data = region_data.round(2)
```
